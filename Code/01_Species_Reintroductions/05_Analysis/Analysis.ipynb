```python
####### Setup
import os
while os.getcwd().split('\\')[-1] != 'ecological-networks':
    %cd ..
import setup_paths
setup_paths.add_path()
from Species_Reintroductions import *
global DATA_TYPE
DATA_TYPE = 'Real_1D'
CASE_TO_SOLVE_init(DATA_TYPE)
get_CASE()

def CASE_init(CASE_TO_SOLVE):
    DATA_TYPE = CASE_TO_SOLVE
    CASE_TO_SOLVE_init(DATA_TYPE)
    NetworkName_Solved_df = UnpickleObj(os.path.join(os.getcwd(), 'Code', '01_Species_Reintroductions', '01_Simulation_Execution', f"Network_Solved_{DATA_TYPE.split('_')[0]}.pkl"))
    M = UnpickleObj(os.path.join(os.getcwd(), 'Code', '01_Species_Reintroductions', '01_Simulation_Execution', f"M_{DATA_TYPE.split('_')[0]}.pkl"))
    if DATA_TYPE.split('_')[0] == 'Syn':
        M = {k:M[k] for k in M.keys() if int(k.split('_')[2]) == 100}
    return DATA_TYPE, NetworkName_Solved_df, M

def init_HeatmapData(Case, Start = 0, End = -1):
    if End == -1:
        End = Start + 1
    if Case == 1:
        Heatmap01_All = UnpickleObj(os.path.join(os.getcwd(), 'Code', '01_Species_Reintroductions', '02_Generate_Database', f"Heatmap01_{DATA_TYPE}.pkl"))
        return {k:Heatmap01_All[k][Start:End] for k in Heatmap01_All.keys()}
    if Case == 2:
        Heatmap02_All = UnpickleObj(os.path.join(os.getcwd(), 'Code', '01_Species_Reintroductions', '02_Generate_Database', f"Heatmap02_{DATA_TYPE}.pkl"))
        return {k:Heatmap02_All[k][Start:End] for k in Heatmap02_All.keys()}
    if Case == 3:
        Heatmap03_All = UnpickleObj(os.path.join(os.getcwd(), 'Code', '01_Species_Reintroductions', '02_Generate_Database', f"Heatmap03_{DATA_TYPE}.pkl"))
        return {k:Heatmap03_All[k][Start:End] for k in Heatmap03_All.keys()}
```

```python
####### Plotting Methods
def Plot_HData_0(C = [0, 0], S = 1, CASE_TO_SOLVE = None, SHOW_PLOT = True):
    
    if CASE_TO_SOLVE != None:
        DATA_TYPE, NetworkName_Solved_df, M = CASE_init(CASE_TO_SOLVE)
    
    if DATA_TYPE.split('_')[1] == '1D':
        pass
    if DATA_TYPE.split('_')[1] == '2D':
        S = S - 1
    
    HData = init_HeatmapData(1, 0)
    Data = []
    for NetworkName in itertools.islice(NetworkName_Solved_df.NetworkName,0,None,1):
        row = []
        for Var_ind in [0,1,2,3,4,5]: ## Variables
            for Strategy in [0,1,2]: ## Perturbation Strategy
                for Fraction in [2,5,8]: ## Fraction of Perturbation
                    row += [HData[NetworkName][0][Var_ind][(Strategy,Fraction,C[0]),C[1]]]
        Data += [row]
    Data = np.array(Data)
    #
    if SHOW_PLOT:
        f,ax = plt.subplots(1,1)
        f.tight_layout(pad=0)
        
        g00 = sns.heatmap(
            Data[:,:,S],
            vmin = 0,
            vmax = 1,
            annot=False,
            annot_kws={"fontsize":8},
            #cmap="rainbow",
            cmap = 'autumn',
            cbar=False,
            linewidth=0.1,
            #linecolor = 'black',
            square = True,
            ax = ax,
            )
        from matplotlib.patches import Rectangle
        ##  Row, Column, Width, Height
        
        ## 0,0,9,30
        Patch = [[0,0,9,len(M.keys())],[9,0,9,len(M.keys())],[18,0,9,len(M.keys())],[27,0,9,len(M.keys())],[36,0,9,len(M.keys())],[45,0,9,len(M.keys())]]
        for var in Patch:
            x, y, w, h = var
            ax.add_patch(Rectangle((x, y), w, h, fill=False, edgecolor='black', lw=3, clip_on=False))
        ax.tick_params(length=0)
        
        g00.set_ylabel('')
        g00.set_xlabel('')
        g00.set_yticks([])
        g00.set_xticks([])

    return Data[:,:,S]

def Plot_HData_1(C = [0, 0], CASE_TO_SOLVE = None, SHOW_PLOT = True):
    
    if CASE_TO_SOLVE != None:
        DATA_TYPE, NetworkName_Solved_df, M = CASE_init(CASE_TO_SOLVE)
    
    HData = init_HeatmapData(1, 1)
    Data = []
    for NetworkName in itertools.islice(NetworkName_Solved_df.NetworkName,0,None,1):
        row = []
        for Var_ind in [0,1,2,3,4,5]: ## Variables
            for Strategy in [0,1,2]: ## Perturbation Strategy
                for Fraction in [2,5,8]: ## Fraction of Perturbation
                    row += [HData[NetworkName][0][Var_ind][(Strategy,Fraction,C[0]),C[1]]]
        Data += [row]
    Data = np.array(Data)
    #
    if SHOW_PLOT:
        f,ax = plt.subplots(1,1)
        f.tight_layout(pad=0)
        
        g00 = sns.heatmap(
            Data,
            vmin = 0,
            vmax = 4,
            annot=False,
            annot_kws={"fontsize":8},
            cmap="rainbow",
            #cmap = 'autumn',
            cbar=False,
            linewidth=0.1,
            #linecolor = 'black',
            square = True,
            ax = ax,
            )
        from matplotlib.patches import Rectangle
        ##  Row, Column, Width, Height
        
        ## 0,0,9,30
        Patch = [[0,0,9,len(M.keys())],[9,0,9,len(M.keys())],[18,0,9,len(M.keys())],[27,0,9,len(M.keys())],[36,0,9,len(M.keys())],[45,0,9,len(M.keys())]]
        for var in Patch:
            x, y, w, h = var
            ax.add_patch(Rectangle((x, y), w, h, fill=False, edgecolor='black', lw=3, clip_on=False))
        ax.tick_params(length=0)
        
        if DATA_TYPE.split('_')[0] == "Syn":
            Patch = [[0,x,Data.shape[1],3] for x in [0,3,6,9,12,15,18,21,24]]
            for var in Patch:
                x, y, w, h = var
                ax.add_patch(Rectangle((x, y), w, h, fill=False, edgecolor='black', lw=1.5, clip_on=False))
            ax.tick_params(length=0)
            g00.set_ylabel(r"$(A,C)$", fontsize = 14)
            g00.set_xlabel('')
            g00.set_yticks(np.arange(27)+0.5)
            g00.set_yticklabels([
                '',r'$(0.25,0.15)$','',
                '',r'$(0.25,0.20)$','',
                '',r'$(0.50,0.10)$','',
                '',r'$(0.50,0.15)$','',
                '',r'$(0.50,0.20)$','',
                '',r'$(0.75,0.05)$','',
                '',r'$(0.75,0.10)$','',
                '',r'$(0.75,0.15)$','',
                '',r'$(0.75,0.20)$',''
                ], fontsize = 8, rotation = 0)
            g00.set_xticks(np.arange(54)+0.5)
            g00.set_xticklabels([
                '','','','',r'<$X$>','','','','',
                '','','','',r'<$ST$>','','','','',
                '','','','',r'<$P$>','','','','',
                '','','','',r'<<$X$>>','','','','',
                '','','','',r'<<$ST$>>','','','','',
                '','','','',r'<<$P$>>','','','',''
                ], fontsize = 12)
        
        if DATA_TYPE.split('_')[0] == "Real":
            '''
            Patch = [[x,0,3,len(M.keys())] for x in [0,3,6,9,12,15,18,21,24,27,30,33,36,39,42,45,48,51]]
            for var in Patch:
                x, y, w, h = var
                ax.add_patch(Rectangle((x, y), w, h, fill=False, edgecolor='green', lw=1.0, clip_on=False))
            ax.tick_params(length=0)
            '''
            g00.set_ylabel(r"$\longleftarrow$ Size of Network", fontsize = 14)
            g00.set_xlabel('')
            g00.set_yticks([])
            g00.set_xticks(np.arange(54)+0.5)
            g00.set_xticklabels([
                '','','','',r'<$X$>','','','','',
                '','','','',r'<$ST$>','','','','',
                '','','','',r'<$P$>','','','','',
                '','','','',r'<<$X$>>','','','','',
                '','','','',r'<<$ST$>>','','','','',
                '','','','',r'<<$P$>>','','','',''
                ], fontsize = 12)
    
    return Data

def Plot_HData_2(C = [0, 0], S = 1, CASE_TO_SOLVE = None, SHOW_PLOT = True):
    
    if CASE_TO_SOLVE != None:
        DATA_TYPE, NetworkName_Solved_df, M = CASE_init(CASE_TO_SOLVE)
    
    if DATA_TYPE.split('_')[1] == '1D':
        pass
    if DATA_TYPE.split('_')[1] == '2D':
        S = S - 1
    
    HData = init_HeatmapData(1, 2)
    Data = []
    for NetworkName in itertools.islice(NetworkName_Solved_df.NetworkName,0,None,1):
        row = []
        for Var_ind in [0,1,2,3,4,5]: ## Variables
            for Strategy in [0,1,2]: ## Perturbation Strategy
                for Fraction in [2,5,8]: ## Fraction of Perturbation
                    row += [HData[NetworkName][0][Var_ind][(Strategy,Fraction,C[0]),C[1]]]
        Data += [row]
    Data = np.array(Data)
    #
    if SHOW_PLOT:
        f,ax = plt.subplots(1,1)
        f.tight_layout(pad=0)
        
        g00 = sns.heatmap(
            #Data,
            Data[:,:,S],
            vmin = 0,
            vmax = 1,
            annot=False,
            annot_kws={"fontsize":8},
            #cmap="rainbow",
            cmap = 'autumn',
            cbar=False,
            linewidth=0.1,
            #linecolor = 'black',
            square = True,
            ax = ax,
            )
        from matplotlib.patches import Rectangle
        ##  Row, Column, Width, Height
        
        ## 0,0,9,30
        Patch = [[0,0,9,len(M.keys())],[9,0,9,len(M.keys())],[18,0,9,len(M.keys())],[27,0,9,len(M.keys())],[36,0,9,len(M.keys())],[45,0,9,len(M.keys())]]
        for var in Patch:
            x, y, w, h = var
            ax.add_patch(Rectangle((x, y), w, h, fill=False, edgecolor='black', lw=3, clip_on=False))
        ax.tick_params(length=0)
        
        g00.set_ylabel('')
        g00.set_xlabel('')
        g00.set_yticks([])
        g00.set_xticks([])

    return Data[:,:,S]

def Plot_HData_3(C = [0, 0], S = 1, V = 0, math = 0, CASE_TO_SOLVE = None, SHOW_PLOT = True):
    
    if CASE_TO_SOLVE != None:
        DATA_TYPE, NetworkName_Solved_df, M = CASE_init(CASE_TO_SOLVE)
    
    if DATA_TYPE.split('_')[1] == '1D':
        pass
    if DATA_TYPE.split('_')[1] == '2D':
        S = S - 1
    
    HData = init_HeatmapData(1, 3)
    Data = []
    
    for NetworkName in itertools.islice(NetworkName_Solved_df.NetworkName,0,None,1):
        row = []
        for Var_ind in [0,1,2,3,4,5]: ## Variables
            for Strategy in [0,1,2]: ## Perturbation Strategy
                for Fraction in [2,5,8]: ## Fraction of Perturbation
                    row += [HData[NetworkName][0][Var_ind][(Strategy,Fraction,C[0]),C[1]]]
        Data += [row]
    Data = np.array(Data)
    #
    if SHOW_PLOT:
        f,ax = plt.subplots(1,1, figsize = (30,30))
        f.tight_layout(pad=0)
        
        g00 = sns.heatmap(
            #Data,
            Data[:,(V*9):9*(V+1), S, math],
            #vmin = 0,
            #vmax = 1,
            annot=True,
            annot_kws={"fontsize":8},
            fmt = 'g',
            #cmap="rainbow",
            #cmap = 'autumn',
            cmap = 'Blues',
            cbar=False,
            linewidth=0.1,
            linecolor = 'black',
            square = True,
            ax = ax,
            )
        '''
        from matplotlib.patches import Rectangle
        ##  Row, Column, Width, Height
        
        Patch = [[0,0,9,30],[9,0,9,30],[18,0,9,30],[27,0,9,30],[36,0,9,30],[45,0,9,30]]
        for var in Patch:
            x, y, w, h = var
            ax.add_patch(Rectangle((x, y), w, h, fill=False, edgecolor='black', lw=3, clip_on=False))
        ax.tick_params(length=0)
        '''
        g00.set_ylabel('')
        g00.set_xlabel('')
        g00.set_yticks([])
        g00.set_xticks([])
    
    return Data[:,(V*9):9*(V+1), S, math]

def Plot_HData_4(C = [0,0], S = 1, V = 0, CASE_TO_SOLVE = None, SPAN = np.arange(9), OUTLIERS = False):
    if CASE_TO_SOLVE != None:
        DATA_TYPE, NetworkName_Solved_df, M = CASE_init(CASE_TO_SOLVE)
    
    if DATA_TYPE.split('_')[1] == '1D':
        pass
    if DATA_TYPE.split('_')[1] == '2D':
        S = S - 1

    if DATA_TYPE.split('_')[1] == '1D':
        pass
    if DATA_TYPE.split('_')[1] == '2D':
        S = S - 1
    
    HData = init_HeatmapData(1, 4)
    Data = []
    for NetworkName in itertools.islice(NetworkName_Solved_df.NetworkName,0,None,1):
        row = []
        for Var_ind in [0,1,2,3,4,5]: ## Variables
            for Strategy in [0,1,2]: ## Perturbation Strategy
                for Fraction in [2,5,8]: ## Fraction of Perturbation
                    row += [HData[NetworkName][0][Var_ind][(Strategy,Fraction,C[0]),C[1]][S,:]]
        Data += [row]
    #Data = np.array(Data)
    del C, row, Fraction, Strategy, Var_ind, NetworkName
    del HData
    
    Data = [X[(V*9):9*(V+1)] for X in Data]
    #
    plt.figure()
    X = []
    for i in range(len(M.keys())):
        Y = []
        for j in SPAN:
            Y += list(Data[i][j])
        X += [Y]
    plt.boxplot(X, showfliers=OUTLIERS, vert = False)
    plt.yticks(ticks = np.arange(1, len(M.keys())+1), labels = [])
    
    return Data
```
