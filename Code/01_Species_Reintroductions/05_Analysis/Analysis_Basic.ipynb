```python
####### Radar Plot
def Plot_Radar(Case = 0):
    ## Radar Chart
    df = UnpickleObj(os.path.join(os.getcwd(), 'Code', '01_Species_Reintroductions', '01_Simulation_Execution', f"Network_Solved_{DATA_TYPE.split('_')[0]}.pkl"))
    if Case == 1:
        df.drop('Size',axis = 1, inplace = True)
    labels = list(df.iloc[:,1:].columns)
    for col in labels:
        df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
    fig=plt.figure()
    ax = fig.add_subplot(111, polar=True)
    for row in range(df.shape[0]):
        stats= df.loc[row,labels].values
        angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
        
        stats=np.concatenate((stats,[stats[0]]))
        angles=np.concatenate((angles,[angles[0]]))
    
        ax.plot(angles, stats,linestyle = 'dotted', alpha = 0.65)
        ax.fill(angles, stats, alpha = 0.35)
    ax.set_yticklabels([])
    ax.set_theta_offset(np.pi*(45/180))
    if Case == 0:ax.set_thetagrids((angles * 180/np.pi)[:-1], ['S','A','C','N'])
    if Case == 1:ax.set_thetagrids((angles * 180/np.pi)[:-1], ['A','C','N'])
    ax.grid(True)
    
    return df.iloc[:,1:].describe()
```

```python
####### Get Masking
def get_Masking(Placeholder = 'Real_1D', C = [0,0], Ref = 'Avg'):
    X0 = Plot_HData_2(C = C, S = 0, CASE_TO_SOLVE=Placeholder,SHOW_PLOT = False);
    X1 = Plot_HData_2(C = C, S = 1, CASE_TO_SOLVE=Placeholder,SHOW_PLOT = False);
    X2 = Plot_HData_2(C = C, S = 2, CASE_TO_SOLVE=Placeholder,SHOW_PLOT = False);
    X3 = Plot_HData_2(C = C, S = 3, CASE_TO_SOLVE=Placeholder,SHOW_PLOT = False);
    X4 = Plot_HData_2(C = C, S = 4, CASE_TO_SOLVE=Placeholder,SHOW_PLOT = False);
    #
    if Ref == 'Avg':
        X0 = [X0[:,0:9],X0[:,9:18],X0[:,18:27]]
        X1 = [X1[:,0:9],X1[:,9:18],X1[:,18:27]]
        X2 = [X2[:,0:9],X2[:,9:18],X2[:,18:27]]
        X3 = [X3[:,0:9],X3[:,9:18],X3[:,18:27]]
        X4 = [X4[:,0:9],X4[:,9:18],X4[:,18:27]]
    elif Ref == 'Var':
        X0 = [X0[:,27:36],X0[:,36:45],X0[:,45:54]]
        X1 = [X1[:,27:36],X1[:,36:45],X1[:,45:54]]
        X2 = [X2[:,27:36],X2[:,36:45],X2[:,45:54]]
        X3 = [X3[:,27:36],X3[:,36:45],X3[:,45:54]]
        X4 = [X4[:,27:36],X4[:,36:45],X4[:,45:54]]
    X = [X0,X1,X2,X3,X4]
    del X0,X1,X2,X3,X4
    Y = [np.ones_like(X[0][0]),np.ones_like(X[0][0]),np.ones_like(X[0][0])]
    Y[0] *= -1
    Y[1] *= -1
    Y[2] *= -1
    #
    for V in [0,1,2]:
        for i in range(X[0][0].shape[0]):
            for j in range(X[0][0].shape[1]):
                temp = [X[S][V][i,j] for S in range(len(X))]
                
                if sum(temp) == 1:
                    Y[V][i,j] = temp.index(1) # 0-4 Strategies
                    
                elif sum(temp) == 5:
                    Y[V][i,j] = 5 # All Strategies
                    
                elif sum(temp) == 2:
                    if (temp[1] == 1) and (temp[2] == 1): # D & C
                        Y[V][i,j] = 6
                    elif (temp[1] == 1) and (temp[3] == 1): # D & B
                        Y[V][i,j] = 7
                    elif (temp[2] == 1) and (temp[3] == 1): # C & B
                        Y[V][i,j] = 8
                    elif temp[4] == 1:
                        temp[4] = 0
                        Y[V][i,j] = temp.index(1)
                    elif temp[0] == 1:
                        Y[V][i,j] = 10 # S + 1 Cen
                        
                elif sum(temp) == 3:
                    if temp[4] == 1:
                        if (temp[1] == 1) and (temp[2] == 1): # D & C
                            Y[V][i,j] = 6
                        elif (temp[1] == 1) and (temp[3] == 1): # D & B
                            Y[V][i,j] = 7
                        elif (temp[2] == 1) and (temp[3] == 1): # C & B
                            Y[V][i,j] = 8
                    elif temp[4] == 0:
                        Y[V][i,j] = 9 # 3 Centralities
                    if Placeholder.split('_')[1] == '2D':
                        if temp[0] == 1:
                            temp[0] = 0
                            Y[V][i,j] = temp.index(1)
                        
                elif sum(temp) == 4:
                    if temp[4] == 1:
                        Y[V][i,j] = 9
                    elif temp[0] == 1:
                        Y[V][i,j] = 5
                        
                else:
                    print(sum(temp))
    #
    # 0-4 Strategies
    # 5 All
    # 6 D & C
    # 7 C & B
    # 8 C & B
    # 9 D & C & B
    # 10 Sys + 1 Cen
    return Y
```

```python
####### % Variation wrt Random
def get_VARwrtRAND(Placeholder = 'Real_1D', V_lst = [0,1,2], C = [0,0], M = 0):
    DATA_TYPE, _, _ = CASE_init(Placeholder)
    if DATA_TYPE.split('_')[1] == '1D':
        S_lst = [0,1,2,3]
    elif DATA_TYPE.split('_')[1] == '2D':
        S_lst = [1,2,3]
    dataset_dict = {}
    for S in S_lst:
        for V in V_lst:
            X = Plot_HData_3(C = C, S = 4, V = V, math = M, CASE_TO_SOLVE=Placeholder, SHOW_PLOT = False);
            Y = Plot_HData_3(C = C, S = S, V = V, math = M, CASE_TO_SOLVE=Placeholder, SHOW_PLOT = False);
    
            X = np.array(X)
            Y = np.array(Y)
            
            dataset_dict[(S,V)] = np.round(((Y - X) / X)*100, 2)
    del S, V, Placeholder
    del X,Y
    
    dataset_describe = pd.DataFrame()
    for V in V_lst:
        for S in S_lst:
            dataset_describe[f"({S},{V})"] = pd.Series(dataset_dict[(S,V)].flatten()).describe()
    del S_lst,S,V
    
    return dataset_describe,dataset_dict
```

```python
####### % Variation wrt some Strategy
def get_VARwrtS(Placeholder = 'Real_1D', V_lst = [0,1,2], SS = [1,4]):
    DATA_TYPE, _, _ = CASE_init(Placeholder)
    dataset_dict = {}
    for V in V_lst:
        X = Plot_HData_3(C = [0,0], S = SS[1], V = V, math = 0, CASE_TO_SOLVE=Placeholder, SHOW_PLOT = False);
        Y = Plot_HData_3(C = [0,0], S = SS[0], V = V, math = 0, CASE_TO_SOLVE=Placeholder, SHOW_PLOT = False);
    
        X = np.array(X)
        Y = np.array(Y)
            
        dataset_dict[(SS[0],V)] = np.round(((Y - X) / X)*100, 2)
    del V, Placeholder
    del X,Y
    
    dataset_describe = pd.DataFrame()
    for V in V_lst:
        dataset_describe[f"({SS[0]}/{SS[1]},{V})"] = pd.Series(dataset_dict[(SS[0],V)].flatten()).describe()
    
    return dataset_describe,dataset_dict
```

```python
def Imp_getDF(DATA_TYPE):
    #DATA_TYPE = 'Real_1D'
    CASE_TO_SOLVE_init(DATA_TYPE)
    M = UnpickleObj(os.path.join(os.getcwd(), 'Code', '01_Species_Reintroductions', '01_Simulation_Execution', f"M_{DATA_TYPE.split('_')[0]}.pkl"))
    if DATA_TYPE.split('_')[1] == '1D':
        loc = '01'
    elif DATA_TYPE.split('_')[1] == '2D':
        loc = '04'
    if DATA_TYPE.split('_')[1] == '1D':
        approach_lst = ['three','four','five','six','rand']
    elif DATA_TYPE.split('_')[1] == '2D':
        approach_lst = ['four','five','six','rand']

    Data = {}
    for NetworkName in itertools.islice(M.keys(),0,None,1):
        DF = UnpickleObj(os.path.join(os.getcwd(), 'Code', '01_Species_Reintroductions', '04_Importance_Data', 'ImportanceData', f"{loc}", f"{NetworkName}.pkl"))
        DF.drop(['N1','X_ob1'], axis = 'columns', inplace = True)
        DF = DF.dropna()
        for col in list(DF.columns):
            DF.drop(list(DF[ DF[col] == np.inf ].index), inplace = True)
        Data[NetworkName] = DF
    del M, loc, DF, NetworkName, col
    #
    for NetworkName in itertools.islice(list(Data.keys()),0,None,1):
        #DF = pd.DataFrame([], columns = ['Approach','Key0','Key1','Key2','S0','A0','C0','N0','X0','X_ob0','S1','A1','C1','N1','X1','X_ob1'])
        DF = pd.DataFrame([], columns = ['Approach','Key0','Key1','Key2','S0','A0','C0','N0','X0','X_ob0','S1','A1','C1','X1'])
        for approach in approach_lst:
            for key in set(list(itertools.product(*[[0,1,2], [2,5,8], [0]]))):
                try:
                    DF_Slice = Data[NetworkName]
                    
                    DF_Slice = DF_Slice[ (DF_Slice['Approach'] == approach) & (DF_Slice['Key0'] == key[0]) & (DF_Slice['Key1'] == key[1]) & (DF_Slice['Key2'] == key[2])]
                    X = np.array(DF_Slice)[:,5:]
                    Y = list(np.mean(X, axis = 0))
                    Y = [approach, key[0], key[1], key[2]] + Y
                    DF = DF.append(pd.Series(Y, index = DF.columns), ignore_index=True)
                except:
                    print(NetworkName, approach, key)
                    Y = [approach, key[0], key[1], key[2]] + [np.nan for i in range(len(DF.columns)-4)]
                    DF = DF.append(pd.Series(Y, index = DF.columns), ignore_index=True)
        DF.sort_values(by = ['Key0','Key1','Key2'], ascending = True, inplace = True)
        Data[NetworkName] = DF
    del NetworkName, approach, key, DF_Slice, X, Y, DF, approach_lst
    return Data

def Imp_Solve(DATA_TYPE, Data, Col = 'X1'):
    DATA_TYPE = DATA_TYPE
    #Data = Imp_getDF(DATA_TYPE)
    if DATA_TYPE.split('_')[1] == '1D':
        approach_lst = ['three','four','five','six','rand']
    elif DATA_TYPE.split('_')[1] == '2D':
        approach_lst = ['four','five','six','rand']
    Z = []
    def Imp_getTemp(Data, approach, Col):
        D = []
        for NetworkName in itertools.islice(list(Data.keys()),0,None,1):
            DF = Data[NetworkName]
            #DF = DF.loc[:,['Approach','Key0','Key1','Key2','X1']]
            row = []
            for key in zip([0,0,0,1,1,1,2,2,2],[2,5,8]*3, [0]*9):
                val = DF[ (DF['Approach'] == approach) & (DF['Key0'] == key[0]) & (DF['Key1'] == key[1]) & (DF['Key2'] == key[2])][Col].values[0]
                row += [val]
            D += [row]
        D = np.array(D)
        return np.mean(D, axis = 0)
    for approach in approach_lst:
        Z += [Imp_getTemp(Data, approach, Col)]
    Z = np.array(Z)
    Z = pd.DataFrame(Z, columns = [key for key in zip([0,0,0,1,1,1,2,2,2],[2,5,8]*3, [0]*9)], index = approach_lst)
    return Z
```
